```python
# Your code goes here
import pandas as pd

url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url3)

```

```python
df1["Customer Lifetime Value"] = pd.to_numeric(df1["Customer Lifetime Value"], errors='coerce')
data_types = df1.dtypes

# Print the data types for each column
print(data_types)
```

```python
df2["Customer Lifetime Value"] = pd.to_numeric(df2["Customer Lifetime Value"], errors='coerce')
df2["Monthly Premium Auto"] = pd.to_numeric(df2["Monthly Premium Auto"], errors='coerce')
data_types = df2.dtypes

# Print the data types for each column
print(data_types)
```

```python
data_types = df3.dtypes
df3['Number of Open Complaints'] = df3['Number of Open Complaints'].astype('object')
# Print the data types for each column
print(data_types)
```

```python
df1.columns = df1.columns.map(lambda x: x.replace(' ', '_').lower())
df1.rename(columns={'st': 'state'}, inplace=True)
print(df1.columns)


df2.columns = df2.columns.map(lambda x: x.replace(' ', '_').lower())
df2.rename(columns={'st': 'state'}, inplace=True)
print(df1.columns)


df3.columns = df3.columns.map(lambda x: x.replace(' ', '_').lower())
df3.rename(columns={'State': 'state'}, inplace=True)
print(df3.columns)
```

```python
df3

```

```python
combined_df = pd.concat([df1, df2, df3], ignore_index=True, axis=0)
combined_df
```

```python
combined_df['gender'] = combined_df['gender'].str.replace('Femal', 'F').str.replace('female', 'F').str.replace('Male', 'M')
combined_df['state'] = combined_df['state'].str.replace('Cali', 'California').str.replace('WA', 'Washington').str.replace('AZ', 'Arizona')
combined_df['education'] = combined_df['education'].str.replace('Bachelors', 'Bachelor')
combined_df['education'] = combined_df['education'].str.replace('Sports Car', 'Luxury').str.replace('Luxury SUV', 'Luxury').str.replace('Luxury Car', 'Luxury')
```

```python
combined_df["customer_lifetime_value"] = pd.to_numeric(combined_df["customer_lifetime_value"], errors='coerce')
combined_df['number_of_open_complaints'] = combined_df['number_of_open_complaints'].str.split('/').str[1]

print(combined_df)
```

```python
combined_df.isna().sum()
```

```python
combined_df = combined_df.dropna(subset=['customer'])
```

```python
combined_df.isnull()
```

```python
combined_df['gender'].fillna(method='ffill', inplace=True)
combined_df['customer_lifetime_value'].fillna(method='ffill', inplace=True)
```

```python
combined_df['gender'].fillna(method='bfill', inplace=True)
combined_df['customer_lifetime_value'].fillna(method='bfill', inplace=True)
```
